### 6 . 텐서플로우를 위한 가상 영역 생성

      conda create -n tensorflow-gpu python=3.6



### 7. 텐서플로우 가상영역 변경 
    
    windows : activate tensorflow-gpu
    
    mac : source activate tensorflow-gpu

### 8. 텐서플로우 가상영역 변경

      deactivate tensorflow-gpu

###  9. 텐서플로우 설치 

     가상환경 내에 설치 
     
      pip 업그레이드 하기 : python -m pip install --upgrade pip
     

      파이썬으로 설치 :  pip3 install --upgrade tensorflow-gpu
      
      
      콘다명령어로 설치 :  conda create -n tensorflow-gpu python=3.6
                         activate tensorflow-gpu
                         conda install -c anaconda tensorflow-gpu

#### pip 명령

        https://suwoni-codelab.com/python%20%EA%B8%B0%EB%B3%B8/2018/03/21/Python-Basic-PIP/

#### conda 명령

    가상환경 내부에 모듈 설치 : conda install --name myenv scipy
    
    특정 버전을 주고 모듈 설치 : conda install scipy=0.15.0 curl=7.26.0
    
    여러 개 모듈을 설치(-c) : conda install -c pandas bottleneck
    
    패키지 삭제 : conda remove -n myenv scipy
    
    가상환경 삭제 : conda remove -n myenv
