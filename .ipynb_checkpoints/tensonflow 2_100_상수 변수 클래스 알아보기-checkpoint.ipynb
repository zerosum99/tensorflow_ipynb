## 변수 클래스 알아보기

```python
dir(tf.Variable)
```

```python
help(tf.Variable)
```

### 변수를 생성하면 하나의 변수 객체가 만들어진다.

```python
a = tf.Variable(0,name="a")
```

```python
a
```

### 변수값을 assign 메소드로 변경이 가능하다.

```python
a.assign(a+1)
```

### 변수의 값을 보면 실제 tensor로 보인다.

```python
a.value()
```

```python
a.read_value()
```
